# Lyric File Cleanup

There are several lyric file's that contain text in brackets. Let's see what they are.

```python
%matplotlib inline
```

```python
import os
import sys
nb_dir = os.path.split(os.getcwd())[0]
project_dir = os.path.join(nb_dir, os.pardir)

if project_dir not in sys.path:
   sys.path.append(project_dir)

from src import corpus
```

```python
import re

raw_dirs = corpus.raw_data_dirs()
raw_songs = corpus.load_songs(raw_dirs[0])
tagged = []
reg = re.compile('\[.*\]')

for song in raw_songs:
    tagged = tagged + reg.findall(song['lyrics'])
    
print(set(tagged))
```



The bracket text can be cleaned out. It doesn't provide any useful information in our analysis.

There is also some text at the top of the file that starts with 'Produced by...'. Let's see if it is useful.

```python
import re

raw_dirs = corpus.raw_data_dirs()
raw_songs = corpus.load_songs(raw_dirs[0])
matches = []
reg = re.compile('[p|P]roduced.*')

for song in raw_songs:
    matches = matches + reg.findall(song['lyrics'])
    
print(set(matches))
```

```python
import re

raw_dirs = corpus.raw_data_dirs()
raw_songs = corpus.load_songs(raw_dirs[0])

tagged = re.compile('\[.*\]')
produced = re.compile('[p|P]roduced.*')

for song in raw_songs:
    new_lyrics = tagged.sub('', song['lyrics'])
    new_lyrics = produced.sub('', new_lyrics)
    print(new_lyrics)
```
